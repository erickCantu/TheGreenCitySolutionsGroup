```python
def run_mlflow_experiment(run_name = None, start = 50):
    mlflow.set_experiment("hello_world_experiment")
    
    with mlflow.start_run(run_name=run_name) as r:
        print("run-uuid", r.info.run_uuid)
        
        building_nr = 5
        mlflow.log_param("testrun", 3)
        mlflow.log_param("building_nr", building_nr)
        mlflow.log_param("start_index", start)
        df = pd.read_csv(f"../data/preprocessed/Building_{building_nr}.csv")
        plt.figure()
        df.iloc[start:start+300, :].outdoor_temp.plot()
        plt.savefig("../mlflow_output/temp_plot.png")
        mlflow.log_artifact("../mlflow_output/")
        #mlflow.end_run() #doesnt the with statement do this?
```
